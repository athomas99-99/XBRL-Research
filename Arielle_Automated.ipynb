***XBRL File Loading***

```python
from arelle import Cntlr, ModelManager,ViewWinFactTable

def load_xbrl_from_url(url):
    # Initialize the Arelle controller
    controller = Cntlr.Cntlr(logFileName="log.txt")

    # Model manager to manage XBRL operations
    model_manager = ModelManager.initialize(controller)

    # Load the XBRL file from URL
    model_xbrl = model_manager.load(url)

    # Print some basic info to verify loading
    if model_xbrl is not None:
        print("XBRL Document Loaded Successfully")
        print("Number of Facts:", len(model_xbrl.facts))
        return model_xbrl
    else:
        print("Failed to load XBRL document")
        return


def load_multiple_xbrl_files(urls):
  for url in urls:
        load_xbrl_from_url(url)

# URL of the XBRL file you want to load
xbrl_urls = xbrl_paths

# Call the function with the specified URL
xbrl_content = load_multiple_xbrl_files(xbrl_urls)
```

***Extracts Linkroles for each XBRL File for Income Statement, Balance Sheet, and Statement of Cash Flow***

```python
import os
from arelle import Cntlr, ModelManager, ModelXbrl

def extract_calculation_relationships(xbrl_file_path):
    """
    Extract calculation relationships from an XBRL file.

    Args:
        xbrl_file_path (str): The path to the XBRL file.

    Returns:
        dict: A dictionary of calculation relationships.
    """
    # Initialize the controller and model manager
    controller = Cntlr.Cntlr(logFileName="log.txt")
    model_manager = ModelManager.initialize(controller)
    model_xbrl = ModelXbrl.load(model_manager, xbrl_file_path)

    # Get the list of all link roles in the XBRL file
    link_roles = model_xbrl.relationshipSet("http://www.xbrl.org/2003/arcrole/parent-child").linkRoleUris

    # Dictionary to store all calculation relationships by link role
    all_calc_relationships = {}

    for link_role in link_roles:
        # Extract the calculation relationships for the current link role
        calc_linkbase = model_xbrl.relationshipSet("http://www.xbrl.org/2003/arcrole/summation-item", linkrole=link_role)
        calc_relationships = {}

        # Store relationships in the dictionary
        for rel in calc_linkbase.modelRelationships:
            from_label = str(rel.fromModelObject.qname)
            to_label = str(rel.toModelObject.qname)
            weight = '+' if rel.weight > 0 else '-'

            if from_label not in calc_relationships:
                calc_relationships[from_label] = []
            calc_relationships[from_label].append((weight, to_label))

        all_calc_relationships[link_role] = calc_relationships

    return all_calc_relationships

# Assuming xbrl_file_paths is a list of paths to XBRL files you already have
xbrl_file_paths = xbrl_paths

IS_tags = [
    "us-gaap:Revenues",
    "us-gaap:CostOfGoodsAndServicesSold",
    "us-gaap:GrossProfit",
    "us-gaap:ProfitLoss",
    "us-gaap:OperatingExpenses",
    "us-gaap:OperatingIncomeLoss"
]

BS_tags = [
    "us-gaap:Assets",
    "us-gaap:Liabilities",
    "us-gaap:StockholdersEquity",
    "us-gaap:Equity",
    "us-gaap:RetainedEarnings"
]

CF_tags = ["us-gaap:NetCashProvidedByUsedInOperatingActivities",
"us-gaap:NetCashProvidedByUsedInInvestingActivities",
"us-gaap:NetCashProvidedByUsedInFinancingActivities",
"us-gaap:DepreciationDepletionAndAmortization",
"us-gaap:AmortizationOfIntangibleAssets",
"us-gaap:PaymentsToAcquirePropertyPlantAndEquipment",
"us-gaap:ProceedsFromSaleOfPropertyPlantAndEquipment",
"us-gaap:CashAndCashEquivalentsPeriodIncreaseDecrease"
]

keywords = [
    "ConsolidatedStatementsOfIncome", "ConsolidatedStatementsOfOperations", "ConsolidatedStatementOfIncome",
    "StatementOfIncome", "IncomeStatement", "ConsolidatedStatementsOfEarnings", "StatementOfEarnings",
    "EarningsStatement", "ProfitAndLossStatement", "StatementOfOperations", "ConsolidatedStatementOfOperations",
    "P&LStatement", "ConsolidatedBalanceSheets", "ConsolidatedBalanceSheet", "BalanceSheet",
    "StatementOfFinancialPosition", "ConsolidatedStatementsOfFinancialPosition", "FinancialPosition",
    "StatementOfAssetsAndLiabilities", "StatementOfCondition", "PositionStatement", "ConsolidatedStatementsOfCashFlows",
    "ConsolidatedStatementOfCashFlows", "StatementOfCashFlows", "CashFlowStatement", "CashFlowsStatement",
    "StatementOfCashFlow", "ConsolidatedCashFlows", "CashFlow", "CONSOLIDATEDSTATEMENTSOFINCOME",
    "CONSOLIDATEDSTATEMENTOFINCOME", "STATEMENTOFINCOME", "INCOMESTATEMENT", "CONSOLIDATEDSTATEMENTSOFEARNINGS",
    "STATEMENTOFEARNINGS", "EARNINGSSTATEMENT", "PROFITANDLOSSSTATEMENT", "STATEMENTOFOPERATIONS",
    "CONSOLIDATEDSTATEMENTSOFOPERATIONS", "CONSOLIDATEDSTATEMENTOFOPERATIONS", "P&LSTATEMENT", "CONSOLIDATEDBALANCESHEETS",
    "CONSOLIDATEDBALANCESHEET", "BALANCESHEET", "STATEMENTOFFINANCIALPOSITION", "CONSOLIDATEDSTATEMENTSOFFINANCIALPOSITION",
    "FINANCIALPOSITION", "STATEMENTOFASSETSANDLIABILITIES", "STATEMENTOFCONDITION", "POSITIONSTATEMENT",
    "CONSOLIDATEDSTATEMENTSOFCASHFLOWS", "CONSOLIDATEDSTATEMENTOFCASHFLOWS", "STATEMENTOFCASHFLOWS",
    "CASHFLOWSTATEMENT", "CASHFLOWSSTATEMENT", "STATEMENTOFCASHFLOW", "CONSOLIDATEDCASHFLOWS", "CASHFLOW"
]

def find_link_role_by_keywords(link_roles, keywords):
    for link_role in link_roles:
        for keyword in keywords:
            if keyword.lower() in link_role.lower():
                return link_role
    return None

def get_primary_link_role(link_roles, keywords):
    if len(link_roles) == 1:
        return link_roles[0]
    elif len(link_roles) > 1:
        # Try to find a link role that matches the keywords
        keyword_matched_role = find_link_role_by_keywords(link_roles, keywords)
        if keyword_matched_role:
            return keyword_matched_role
    else:
        return None

# Dictionaries to store the link roles for each financial statement type
IS_dict = {}
BS_dict = {}
CF_dict = {}
# Iterate over the list of XBRL file paths and process each file
for xbrl_url in xbrl_file_paths:
    print(f"\nProcessing file: {xbrl_url}")
    all_relationships = extract_calculation_relationships(xbrl_url)

    IS_linkRole = []
    BS_linkRole = []
    CF_linkRole = []

    # Determine the link roles for each financial statement
    for link_role, relationships in all_relationships.items():
        if any(tag in relationships for tag in IS_tags):
            IS_linkRole.append(link_role)
        elif any(tag in relationships for tag in BS_tags):
            BS_linkRole.append(link_role)
        elif any(tag in relationships for tag in CF_tags):
            CF_linkRole.append(link_role)

    IS_primary = get_primary_link_role(IS_linkRole, keywords)
    BS_primary = get_primary_link_role(BS_linkRole, keywords)
    CF_primary = get_primary_link_role(CF_linkRole, keywords)

    if IS_primary:
        IS_dict.update({xbrl_url: IS_primary})
        print(f"\n1. Income Statement Link Role: {IS_primary}")
    else:
        print(f"\n1. Error: Incorrect Number of link roles for income statement: {IS_linkRole}")
    if BS_primary:
        BS_dict.update({xbrl_url: BS_primary})
        print(f"\n2. Balance Sheet Link Role: {BS_primary}")
    else:
        print(f"\n2. Error: Incorrect Number of link roles for balance sheet: {BS_linkRole}")
    if CF_primary:
        CF_dict.update({xbrl_url: CF_primary})
        print(f"\n3. Cash Flow Statement Link Role: {CF_primary}")
    else:
        print(f"\n3. Error: Incorrect Number of link roles for cash flow statement: {CF_linkRole}")

```
